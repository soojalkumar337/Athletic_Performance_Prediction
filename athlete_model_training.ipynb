# Athletic Performance Prediction

Synthetic dataset generation, preprocessing, model training, and evaluation.


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import joblib
```

```python
df = pd.read_csv('athlete_data.csv')
df.head()
```

```python
# Preprocess

df2 = df.copy()
df2['gender'] = df2['gender'].map({'Male':1,'Female':0})
X = df2.drop(columns=['performance_score','performance_level'])
y = df2['performance_level']
le = LabelEncoder()
y_enc = le.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X, y_enc, test_size=0.2, random_state=42, stratify=y_enc)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
model = RandomForestClassifier(n_estimators=300, random_state=42)
model.fit(X_train_s, y_train)
y_pred = model.predict(X_test_s)
print('Accuracy:', accuracy_score(y_test, y_pred))
print('\nClassification report:\n', classification_report(y_test, y_pred, target_names=le.classes_))
joblib.dump(model,'athlete_model.pkl')
joblib.dump(scaler,'scaler.pkl')
joblib.dump(le,'label_encoder.pkl')
```
